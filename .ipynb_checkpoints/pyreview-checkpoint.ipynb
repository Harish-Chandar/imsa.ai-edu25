```python
# if statement
if z:
    print("Z IS TRUE")
else:
    print("Z MUST BE FALSE")

if not z:
    print("Z IS NOT TRUE")
else:
    print("Z IS NOT FALSE")
```

```python
def print_sum(x, y):
    print(x + y)

print_sum(34,5)
print_sum(109,10)
```

```python
def return_sum(x, y):
    return(x + y)

print(return_sum(34,5))
print(return_sum(109,10))
```
